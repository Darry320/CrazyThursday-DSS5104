```python
import pandas as pd
import numpy as np
import time
import psutil
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# è®¾ç½®æ•°æ®è·¯å¾„
file_path = r"C:\Users\å¼ å‡¤æ™º\Downloads\higgs\HIGGS.csv.gz"

def evaluate_model(X, y):
        
    # è®­ç»ƒé›† / æµ‹è¯•é›†åˆ’åˆ†
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=999)

    # å¼€å§‹æ—¶é—´è®°å½•
    start_time = time.time()
    process = psutil.Process()

    # æ¨¡å‹åˆå§‹åŒ–å’Œè®­ç»ƒ
    model = RandomForestClassifier(
        n_estimators=100,   # æ ‘çš„æ•°é‡
        n_jobs=-1,          # å¹¶è¡ŒåŠ é€Ÿ
        random_state=999
    )
    
    model.fit(X_train, y_train)

    # ç»“æŸæ—¶é—´è®°å½•
    end_time = time.time()
    elapsed_time = end_time - start_time
    cpu_percent = process.cpu_percent(interval=0.1)
    memory_used = process.memory_info().rss / (1024 ** 2)  # è½¬æ¢ä¸º MB

    # æ¨¡å‹é¢„æµ‹
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    # è¯„ä¼°æŒ‡æ ‡
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)

    print(f"âœ… Accuracy: {acc:.4f}")
    print(f"âœ… F1-score: {f1:.4f}")
    print(f"âœ… AUC: {auc:.4f}")
    print(f"â±ï¸ Time: {elapsed_time:.2f} seconds")
    print(f"ğŸ§  CPU Usage: {cpu_percent}%")
    print(f"ğŸ§  Memory Used: {memory_used:.2f} MB")
```

```python
# åŠ è½½æ•°æ®ï¼ˆä»…å‰ 1,000,000 è¡Œå¯æé«˜é€Ÿåº¦ï¼Œä¹Ÿå¯ä»¥åŠ è½½å…¨éƒ¨ï¼‰
print("ğŸ“¥ æ­£åœ¨åŠ è½½æ•°æ®...")
df = pd.read_csv(file_path, compression='gzip', header=None)

# ç¬¬ä¸€åˆ—æ˜¯æ ‡ç­¾ï¼Œåç»­æ˜¯ç‰¹å¾
y = df.iloc[:, 0]
X = df.iloc[:, 1:]


evaluate_model(X, y)
```
